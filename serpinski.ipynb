```python
def lines_shift(s, n):
	return " "*n + s.replace("\n", "\n" + " "*n)

def lines_double(s):
	slines = s.split("\n")
	return "\n".join([sline + " " + sline for sline in slines])

def regularize(s):
    slines = s.split("\n")
    maxlen = max([len(sline) for sline in slines])
    return "\n".join([sline + " "*(maxlen-len(sline)) for sline in slines])

def next_triangle(t):
    n = len(t.split("\n"))
    return regularize(lines_shift(t, n) + "\n" + lines_double(t))
```
